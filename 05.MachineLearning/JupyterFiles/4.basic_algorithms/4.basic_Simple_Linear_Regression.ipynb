# 0. 라이브러리 불러오기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings(action='ignore')
%config InlineBackend.figure_format='retina'
```

# 1. 데이터 불러오기

```python
# 데이터 읽어오기
path = '../data/simple_car_data.csv'
data = pd.read_csv(path)
```

# 2. 데이터 확인하기

```python
# 상위 몇 개 행 확인
data.head()
```

```python
# 기술통계 확인
data.describe()
```

```python
# NaN 값 확인
data.isnull().sum()
```

```python
# speed, dist 관계
plt.scatter(x='speed', y='dist', data=data)
plt.xlabel('Speed(mph)')
plt.ylabel('Dist(ft)')
plt.show()
```

# 3. 데이터 준비하기

**1) x, y 분리**

- target 변수를 명확히 지정합니다.
- target을 제외한 변수를 x 데이터프레임으로 선언합니다.
- y 데이터프레임은 target 변수만을 가집니다.

```python
# target 확인
target = 'dist'

# 데이터 분리
x = data.drop(target, axis=1)
y = data.loc[:, target]
```

**2) 학습용, 평가용 데이터 분리**

- 학습용, 평가용 데이터를 적절한 비율로 분리합니다.
- 반복 실행 시 동일한 결과를 얻기 위해 random_state 옵션을 지정합니다.

```python
# 모듈 불러오기
from sklearn.model_selection import train_test_split

# 7:3으로 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

# 4. 모델링

```python
# 1단계: 불러오기
from sklearn.linear_model import LinearRegression
```

```python
# 2단계: 선언하기
model = LinearRegression()
```

```python
# 3단계: 학습하기
model.fit(x_train, y_train)
```

```python
# 4단계: 예측하기
y_pred = model.predict(x_test)
```

```python
# 5단계: 평가하기
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test, y_pred))
```

### 회귀 계수 살펴보기

```python
# 회귀계수 확인
print(list(x_train))
print(model.coef_) # 회귀계수(=가중치) 확인
print(model.intercept_) # 편향 수치 확인


# 회귀식 만들기 y = ax + b
a = model.coef_
b = model.intercept_
speed = np.linspace(x_test.min(), x_test.max(),5)
dist = a*speed + b
```

$$ \large Distance = 3.91 * Speed -16.37$$

# 5. 시각화

```python
# 시각화
plt.plot(y_test.values, label='Actual')
plt.plot(y_pred, label='Predicted')
plt.legend()
plt.ylabel('Dist(ft)')
plt.show()
```

```python
plt.scatter(x_test['speed'], y_test.values, label='Scatter')
plt.plot(speed, dist, color='r', label="Linear Regression Line")

plt.legend()
plt.show()
```
