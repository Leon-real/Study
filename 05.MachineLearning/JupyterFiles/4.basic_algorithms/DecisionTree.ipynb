# 2. 데이터 전처리

1. 변수 제거
2. x,y 분리
3. 가변수화
4. 학습용, 평가용 데이터 분리

**1) 변수 제거**

```python
# 제거 대상: id
drop_cols = ['id']

# 변수 제거
data.drop(drop_cols, axis=1, inplace=True)

# 확인
data.head()
```

**2) x, y 분리**

```python
# Target 설정
target = 'CHURN'

# 데이터 분리
x = data.drop(target, axis=1)
y = data.loc[:, target]

x.head()
```

**3) 가변수화**

```python
# 가변수화 대상: REPORTED_SATISFACTION, REPORTED_USAGE_LEVEL, CONSIDERING_CHANGE_OF_PLAN
dumn_cols=['REPORTED_SATISFACTION', 'REPORTED_USAGE_LEVEL', 'CONSIDERING_CHANGE_OF_PLAN']

# 가변수화
x = pd.get_dummies(x, columns=dumn_cols, drop_first=True)

# 확인
x.head()
```

**4) 학습용, 평가용 데이터 분리**

```python
# 모듈 불러오기
from sklearn.model_selection import train_test_split

# 7:3으로 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

# 3. 모델링

```python
# 모델 라이브러리 불러오기
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# 모델 선언하기
model = DecisionTreeClassifier(random_state=1, max_depth=5)
```

```python
# 데이터 학습하기
model.fit(x_train, y_train)
```

```python
# 데이터 예측하기
y_pred = model.predict(x_test)

print(y_pred[:3])
print(y_test[:3].values)
```

```python
# 학습된 데이터 평가하기
print(classification_report(y_test, y_pred))
```
