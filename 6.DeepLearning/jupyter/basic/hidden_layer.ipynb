## 0. 라이브러리 및 데이터 준비

```python
import pandas as pd
import tensorflow as tf

# 데이터를 준비합니다.
path = "../../data/boston.csv"
df = pd.read_csv(path)
df.head(2)
```

## 1. x,y 분리

```python
x_train = df[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
y_train = df[['medv']]
print(x_train.shape, y_train.shape)
```

## 2. 모델 준비

```python
# 모델을 준비합니다.
X = tf.keras.Input(shape=[13])
H = tf.keras.layers.Dense(16, activation='swish')(X)
H = tf.keras.layers.Dense(16, activation='swish')(H)
Y = tf.keras.layers.Dense(1)(H)

model = tf.keras.Model(X, Y)
model.compile(loss='mse')
model.summary()
```

## 3. 모델 학습하기

```python
# 모델 학습
model.fit(x_train, y_train, epochs=1000, verbose=0)
model.fit(x_train, y_train, epochs=10)
```

## 4. 모델 이용하기

```python
# 모델을 이용합니다.
y_pred = model.predict(x_train[0:5])
print(y_pred)
```
