### 모집단과 표본

표본을 뽑는 목적 → 모집단 추정하기 위해서

- 표본을 가지고 어떤 통계량을 계산한다면, 그 목적은 모집단을 추정하기 위함이다.
- **표본 평균**
    - 모 평균에 대한 추정치
    - 추정치에는 오차가 존재, 이 오차를 **표준 오차**
    

### 중심극한정리(Central Limit Theorem)

“*평균 μ , 표준편차 σ를 가지는 모집단 분포에서 iid 한 표본을 충분히 많이 추출한다면, 표본 평균은 정규분포에 근사하게 된다 .”*

- 중심 극한 정리가 성립되는 조건을 자세히 명시 하자면,
    1. 유한한 평균과 표준편차를 가지는 모집단 분포
    2. 그 모집단 분포에서, 충분히 많은 표본의 추출
    3. 각각의 표본들은 독립항등분포

### 표준오차

- 모평균 : 추정하고 싶은 정답
- 표본평균 : 모평균에 대한 추정치(예측치)
- 모평균과 표본 평균의 차이 ⇒ 오차 error
- 표본평균 분포의 표준 편차 ⇒ 표준오차

### 95% 신뢰 구간

- 표준오차를 바탕으로 95% 확률 구간을 구할 수 있다.
    - 표본평균으로 모평균을 추정할 때, 95%의 신뢰 구간의 의미는
        - 신뢰 구간 안에 모평균이 포함될 확률이 95%
        - 표본을 100번 정도 뽑으면 95번 정도는 95% 신뢰 구간 안에 모평균을 포함한다.
