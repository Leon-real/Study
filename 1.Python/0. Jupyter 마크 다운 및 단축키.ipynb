# Jupyter notebook 사용법

##마크다운은 아래와 같다.

- Header # : "#"의 개수로 header의 크기를 조절할 수 있다(1~6개까지)
```
# 헤더입니다.
```

- Quote > : ">"을 사용하여 인용문을 만들수 있다.
```
> 인용문입니다.
```

- List : 순서가 있을 경우, 숫자로 표현하고, 순서가 없을경우 "+","*","-"로 표현할 수 있다.
```
1. 첫번째
2. 두번째

- 리스트입니다.
```

- 수평선 : "---' 또는 " *** "로 사용할 수 있다
```
---
```

- 줄바꿈 : 띄어쓰기 2번 또는 <br>을 써준다
```
줄바꿈입니다.<br>오호
```

- 글꼴 굵기, 기울기 : " ** "는 굵기를, " * " 는 기울기를 나타낼 수 있다.
```
**글꼴의 굵기가 굵어져요**
*글꼴의 기울기가 기울어져요*
```

- Link : 주소를 그냥 넣어주면 된다. 만약 지정해준 글자에 링크 넣고 싶다면 아래 예시와 같다.
```
[링크이름](주소)
```

- Image : 아래 예시와 같다.
```
![이미지 이름](이미지 주소)
```

- code : 아래 예시와 같다
```
 ```코드 열기
 여기 코드 작성
 ```코드 닫기
```

- Tex문법 및 수식 넣기 : " $ " 사이에 글을 적으면 를 이용하면 된다
```
$y=3x$
```

## 단축키로는 아래와 같다.

##### 셀 실행, 이동
- Shift + Enter : 셀 실행, 다음 셀 이동
- Ctrl + Enter : 셀 실행
- Alt + Enter : 셀 실행, 아래 새 셀 생성
---

##### 셀 생성, 셀 삭제, 저장
- A : 현재 셀 위에 빈 셀 추가
- B : 현재 셀 아래에 빈 셀 추가
- D, D : 현재 셀 삭제
- Z : 셀 삭제 취소
- S : 저장
---
- X : 현재 셀 잘라내기
- C : 현재 셀 복사
- V : 현재 셀 아래에 붙여 넣기
---
##### 셀 전환
- Y : 코드 셀로 전환
- M : 마크다운 셀로 전환
---

##### 셀 합치기 및 분할
- Shift + M : 아래셀과 합치기
- Ctrl + Shift +- : 셀 분할
---

##### 기타
- Shift + Tab : 툴팁 표시
- L : 코드 셀에서 라인번호 표시
- Ctrl+/ : 주석
